```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('tested.csv')
data
```

```python
data.info()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean()) 
data.info()
```

```python
data = data.drop(columns=['Cabin'])
data.info()
```

```python
data['Fare'] = data['Fare'].fillna(data['Fare'].mean())
data.info()
```

```python
data['Age'] = data['Age'].astype(int)
data.info()
```

```python
pivot_table = pd.pivot_table(data,values='Age',index='PassengerId',columns='Sex')
pivot_table
```

```python
arr2 = np.array(data['Sex'].value_counts())
print(type(arr2))
print(arr2)
print(data['Sex'].value_counts())
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
arr = ['Male','Female']
ax.pie(arr2,labels=arr,autopct='%1.2f%%')
plt.legend(title='Total Population = 418')
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.bar(arr,arr2)
plt.show()
```
