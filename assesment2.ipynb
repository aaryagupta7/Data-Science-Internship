```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

data = pd.read_csv('Heart_Disease_Prediction.csv')
data
```

```python
data.info()
```

Bar Graph

```python
x = data['Age']
y = data['BP']
plt.bar(x,y)
plt.xlabel('Age')
plt.ylabel('BP')
plt.title('Bar Graph')
plt.show()
```

Scatter Plot

```python
x = data['BP']
y = data['Cholesterol']
plt.scatter(x,y)
plt.xlabel('BP')
plt.ylabel('Cholesterol')
plt.title('Scatter Plot')
plt.show()
```

Stacked Bar Chart


```python
x = data['Age']
y1 = data['BP']
y2 = data['Cholesterol']
plt.bar(x,y1,label='y1')
plt.bar(x,y2,label='y2',bottom=y1)
plt.xlabel('Age')
plt.ylabel('BP + Cholesterol')
plt.title('Stacked Bar Chart')
plt.show()
```

Pie Chart

```python
arr = data['Sex'].value_counts()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(arr,labels=['Male','Female'],autopct='%1.2f%%')
plt.legend(title='Total = 270')
plt.show()
```
